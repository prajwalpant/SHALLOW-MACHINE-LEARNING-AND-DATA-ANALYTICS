# Applying Algorithms

```python
 from sklearn.preprocessing import MinMaxScaler
 min_max=MinMaxScaler()
# Scaling down both train and test data set
 X_train_minmax=min_max.fit_transform(train)
 test_min_max=MinMaxScaler()
 X_test_minmax=min_max.fit_transform(test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=15, shuffle=True, random_state=0)
```

# Applying Logistic Regression

```python
clf = LogisticRegression(random_state = 0)
scoring = 'accuracy'
score = cross_val_score(clf,  X_train_minmax, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

#  Applying Knn Algorithm

```python
clf = KNeighborsClassifier(n_neighbors = 10)
scoring = 'accuracy'
score = cross_val_score(clf,  X_train_minmax, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

# Applying Random Forest Algorithm

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf,  X_train_minmax, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

# Applying Decision Tree

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, X_train_minmax, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

# Applying Naive Bayes
